```python
#Vamos a proceder a trabajar con una copia 

df1 = dfflight.copy()
df2 = dfloyalty.copy()
```

```python
df1.columns = df1.columns.str.lower()
df2.columns = df2.columns.str.lower()
```

```python
df2.sample()
```

```python
def reemplazar_guion_bajo(df):
    """
    Reemplaza los espacios en los nombres de columnas por guiones bajos usando un bucle.
    """
    nuevos_nombres = {}

    for col in df.columns:
        nuevos_nombres[col] = col.replace(' ', '_')

    df.rename(columns=nuevos_nombres, inplace=True)
    return df
```

```python
df1 = reemplazar_guion_bajo(df1)
df2 = reemplazar_guion_bajo(df2)
```

```python
print(df1.sample(3))
print(df2.sample(3))
```

```python
df1 = df1.rename(columns={'loyalty_number': 'frequency_id'})

df1['frequency_id'].head()
```

```python
df1.insert(0, 'client_id', df1['frequency_id'])
```

```python
df1.sample(3)
```

```python
# Agrupamos y sumamos valores numéricos, contamos la frecuencia en frequency_id
df1_1 = df1.groupby('client_id').agg({
    'frequency_id': 'count',  # las apariciones de frequency_id por client_id
    'flights_booked': 'sum',
    'flights_with_companions': 'sum',
    'distance': 'sum',
    'points_accumulated': 'sum',
    'points_redeemed': 'sum',
    'dollar_cost_points_redeemed': 'sum',
    'year': ['min', 'max', lambda x: x.nunique()],
}).reset_index()

# Aplanar el MultiIndex
df1_1.columns = [
    'client_id',
    'frequency_count',  # Número de veces que aparece frequency_id por client_id
    'total_flights_booked',
    'total_flights_with_companions',
    'total_distance',
    'total_points_accumulated',
    'total_points_redeemed',
    'total_dollar_cost_redeemed',
    'first_year',
    'last_year',
    'active_years'
]

# Calcular el mes más frecuente
favorite_month = df1.groupby('client_id')['month'].agg(lambda x: x.mode()[0] if not x.mode().empty else None)
favorite_month.name = 'favorite_month'
favorite_month = favorite_month.reset_index()

# Unir con merge
df1_1 = df1_1.merge(favorite_month, on='client_id', how='left')

```

```python
df1_1.rename(columns={'client_id': 'loyalty_number'}, inplace=True)
```

```python
df_merged = pd.merge(
    df2,                  # DataFrame 2 con ID único de clientes en común
    df1_1,                # DataFrame 1
    on='loyalty_number',  # Columna común (ID)
    how='left'            
)
```

```python
df_merged.to_csv('csv/flight-loyalty-activity.csv', index=False)
```
